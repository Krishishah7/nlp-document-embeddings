```python
import numpy as np
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sentences = [
    ["machine", "learning", "is", "fun"],
    ["deep", "learning", "uses", "neural", "networks"],
    ["nlp", "deals", "with", "text", "data"],
    ["word", "embeddings", "capture", "meaning"]
]
```

```python
model = Word2Vec(
    sentences,
    vector_size=50,
    window=3,
    min_count=1,
    workers=4
)
```

```python
def document_embedding(sentence, model):
    vectors = []
    for word in sentence:
        if word in model.wv:
            vectors.append(model.wv[word])
    return np.mean(vectors, axis=0)
```

```python
doc_vectors = []

for sentence in sentences:
    doc_vectors.append(document_embedding(sentence, model))
```

```python
similarity = cosine_similarity(
    [doc_vectors[0]],
    [doc_vectors[1]]
)

print("Document Similarity:", similarity[0][0])
```

How to Interpret the Similarity Score

Closer to 1 → very similar meaning

Closer to 0 → weak similarity

Negative → opposite meaning (rare here)
